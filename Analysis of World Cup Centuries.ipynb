# **ANALYSIS OF TOP RUN SCORERS IN WORLD CUP**

---


```python
# Group the data by player and calculate the total runs
player_runs = data.groupby('Player')['Runs'].sum()

# Sort the players based on their total runs in descending order
top_run_scorers = player_runs.sort_values(ascending=False)

# Print the top run-scorers
print("Top Run-scorers in World Cup centuries:")
print(top_run_scorers.head())
```

## **COMPARISON OF SHOT DISTRIBUTION(4s AND 6s) AGAINST DIFFERENT TEAMS**

---

```python
# Group the data by opposition and calculate the total 4s and 6s
opposition_shots = data.groupby('Opposition')[['4s', '6s']].sum()

# Plot the comparison of shot distribution against different opposition teams
opposition_shots.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Opposition Team')
plt.ylabel('Count')
plt.title('Comparison of Shot Distribution against Different Teams')
plt.show()
```

## **ANALYSIS OF TEAM PERFORMANCE IN LORD'S, LONDON**

---

```python
# Group the data by team and count the total number of centuries
team_centuries = data.groupby('Team').size()

# Group the data by team and venue and count the number of centuries
team_venue_centuries = data.groupby(['Team', 'Venue']).size().reset_index(name='Centuries')

# Select a specific venue for analysis
selected_venue = "Lord's, London"

# Filter the data for the selected venue
venue_data = team_venue_centuries[team_venue_centuries['Venue'] == selected_venue]

# Sort the data by the number of centuries in descending order
venue_data_sorted = venue_data.sort_values('Centuries', ascending=False)

# Plot the team performance in the selected venue
plt.figure(figsize=(10, 6))
plt.bar(venue_data_sorted['Team'], venue_data_sorted['Centuries'])
plt.xlabel('Team')
plt.ylabel('Number of Centuries')
plt.title('Team Performance in ' + selected_venue)
plt.xticks(rotation=45)
plt.show()
```

## **COMPARISON OF CENTURIES SCORED BY TEAMS AGAINST DIFFERENT OPPOSITIONS**

---

```python
# Group the data by team and opposition, and count the number of centuries
team_opposition_centuries = data.groupby(['Team', 'Opposition']).size()

# Print the comparison of centuries scored by teams against different oppositions
print("Comparison of Centuries Scored by Teams against Different Oppositions:")
print(team_opposition_centuries)
```

## **RELATIONSHIP BETWEEN BATTING AVERAGE AND STRIKE RATE IN LORD'S, LONDON**

---

```python
# Select a specific venue for analysis
selected_venue = "Lord's, London"

# Filter the data for the selected venue
venue_data = data[data['Venue'] == selected_venue]

# Group the data by player and calculate the batting average and strike rate
player_stats = venue_data.groupby('Player').agg({'Runs': 'mean', 'S/R': 'mean'}).reset_index()

# Plot the relationship between batting averages and strike rates
plt.figure(figsize=(8, 6))
plt.scatter(player_stats['Runs'], player_stats['S/R'])
plt.xlabel('Batting Average')
plt.ylabel('Strike Rate')
plt.title('Relationship between Batting Average and Strike Rate in ' + selected_venue)
plt.show()
```

## **BATTING AVERAGE AND STRIKE RATE COMPARISON: STRONG VS. WEAK OPPOSITIONS**

---

```python
# Define the criteria to identify strong or weak opposition teams
strong_teams = ['Australia', 'India', 'South Africa', 'England']
weak_teams = ['Zimbabwe', 'Kenya', 'Canada', 'Netherlands']

# Create a new column to categorize the opposition team as strong or weak
data['Opposition Category'] = data['Opposition'].apply(lambda x: 'Strong' if x in strong_teams else 'Weak')

# Calculate the batting average for each player against strong and weak oppositions
batting_average = data.groupby(['Player', 'Opposition Category'])['Runs'].mean().reset_index()

# Calculate the strike rate for each player against strong and weak oppositions
strike_rate = data.groupby(['Player', 'Opposition Category'])['S/R'].mean().reset_index()

# Plot the batting average and strike rate comparison
plt.figure(figsize=(24, 8))
plt.subplot(1, 2, 1)
plt.bar(batting_average['Player'], batting_average['Runs'], color=['#1f77b4', '#ff7f0e'])
plt.xlabel('Player')
plt.ylabel('Batting Average')
plt.title('Batting Average Comparison')
plt.xticks(rotation=45)
plt.xticks(fontsize=10)

plt.subplot(1, 2, 2)
plt.bar(strike_rate['Player'], strike_rate['S/R'], color=['#1f77b4', '#ff7f0e'])
plt.xlabel('Player')
plt.ylabel('Strike Rate')
plt.title('Strike Rate Comparison')
plt.xticks(rotation=45)
plt.xticks(fontsize=10)
plt.tight_layout()
plt.show()
```

#####Here Australia, India, South Africa and England is assumed to be strong teams and Zimbabwe, Kenya, Canada and Netherlands is assumed to be weak teams
